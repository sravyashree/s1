```python
def add(a,b):
    print(a+b)
def sub(a,b):
    print(a-b)
def mul(a,b):
    print(a*b)
def div(a,b):
    print(a/b)

def main():
    num=eval(input("enter a number:"))
    a=eval(input("enter 1st number:"))
    b=eval(input("enter 2nd number:"))
    if num==1:
        add(a,b)
    elif num==2:
        sub(a,b)
    elif num==3:
        mul(a,b)
    elif num==4:
        div(a,b)
```
